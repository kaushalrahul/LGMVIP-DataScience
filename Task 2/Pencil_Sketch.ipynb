**Image to Pencil Sketch**

```python
# Importing the libraries

import matplotlib.pyplot as plt
import cv2
```

**Importing the image from directory**

```python
img = cv2.imread('lotus.jpg')
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap='gray')
plt.show()
```

**Grayscaling Image**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(gray_img, cmap='gray')
plt.title('Gray Image')
plt.show()
```

**Inverting the grayscale image**

```python
invert_img = 255 - gray_img
plt.figure(figsize=(10, 10))
plt.imshow(invert_img, cmap='gray')
plt.title("Inverted Image")
plt.show()
```

**Blurring the inverted image**

```python
blur = cv2.GaussianBlur(invert_img, (93, 93), 0)
plt.figure(figsize=(10, 10))
plt.imshow(blur, cmap='gray')
plt.title("Blurred Inverted Image")
plt.show()
```

**Sketch of the image**

```python
inverted_blur = 255 - blur
ps = cv2.divide(gray_img, inverted_blur, scale=256)
plt.figure(figsize=(10, 10))
plt.imshow(ps, cmap='gray')
plt.title("Pencil Sketch")
plt.show()
```
